# plotting map data in R

```r
library(tidyverse)
```

```r
df = read_csv("data/beer-loc-small.csv")
```

# plotting the continental united states

```r
states = map_data("state")
usa_plot = ggplot(data = states) + 
  geom_polygon(aes(x = long, y = lat, fill = region, group = group), color = "white") + 
  coord_fixed(1.3) +
  guides(fill=FALSE)  # do this to leave off the color legend
```

```r
usa_plot
```

```r
brewery_plot = usa_plot +
  geom_point(data = df, aes(x = long, y = lat), color = "black", size = 2)

brewery_plot
```
